### Проверка расчета эффективности обогащения на обучающей выборке для признака "rougher.output.recovery"

Эффективность обогащения расчитывается по формуле:

**C * (F - T) * 100% / (F * (C - T))**

где:
- C — доля золота в концентрате после флотации/очистки;
- F — доля золота в сырье/концентрате до флотации/очистки;
- T — доля золота в отвальных хвостах после флотации/очистки.

Вычислим эффективность обогащения для признака "rougher.output.recovery". Найдем MAE между нашими расчётами и значением признака. Для начала сохраним необходимые данные в отдельные переменные

```python
check_rec = train[['rougher.input.feed_au', 'rougher.output.concentrate_au',\
                       'rougher.output.tail_au', 'rougher.output.recovery']]
C = check_rec['rougher.output.concentrate_au']
F = check_rec['rougher.input.feed_au']
T = check_rec['rougher.output.tail_au']
```

Добавим столбец с нашими расчетами эффективности

```python
check_rec['my_rec'] = 100 * C * (F - T) / (F * (C - T))
```

Заменим все получившиеся аномальные значения на NAN, после этого в таблице оставим только два столбца с итогами эффективности и заменим все nan-ы на 0. Выведем первые 5 строк для проверки

```python
check_rec.loc[(check_rec['my_rec'] > 100) | (check_rec['my_rec'] < 0), 'my_rec'] = np.nan
check_rec = check_rec[['rougher.output.recovery', 'my_rec']].fillna(0)
display(check_rec.head())
```

Проверим различие между расчетами

```python
check_rec_mae = mean_absolute_error(check_rec['rougher.output.recovery'], check_rec['my_rec'])
print(f'MAE: {check_rec_mae}')
```

Кроме того проверим влияние масштабирования признаков на качество предсказаний. В качестве метрики качества используем sMAPE:

sMAPE = (1/n) * Σ(|target – predict| / ((|target| + |predict|)/2) * 100

В качестве проверочных моделей используем линейную регрессию, дерево решений и регрессию Хубера. Предсказывать будем  эффективность обогащения чернового концентрата на обучающей выборке по следующим признакам:

- 'rougher.state.floatbank10_f_level'

Напишим функция для расчета метрики качества модели sMAPE

```python
def sMAPE(predict, target):
    return 1/len(target) * np.sum(np.abs(target - predict) * 200 / (np.abs(target) + np.abs(predict)))
```

Добавим функцию, которая будет проверять на определенных моделях определенные способы обработки пропусков и возвращать наименование метода и получившееся значение sMAPE

```python
# info - наименование метода
# trans - использовать SimpleImputer() для обработки пропусков или нет?
# strat - определение стратегии для SimpleImputer()
# standard - использовать масштабирование признаков или нет?
def check_on_models(df, info, trans, strat, standard, regressor):
    x_t, x_v, y_t, y_v = train_test_split(df.drop('rougher.output.recovery', axis=1),\
                                                          df['rougher.output.recovery'],\
                                                          test_size=.25, random_state=RAST)
    if trans:
        imp_mean = SimpleImputer(strategy=strat, copy=False)
        imp_mean.fit(x_t)
        imp_mean.transform(x_t)
        imp_mean.transform(x_v)
    if standard:
        numeric = x_t.columns
        scaler = StandardScaler()
        scaler.fit(x_t[numeric])
        x_t[numeric] = scaler.transform(x_t[numeric])
        x_v[numeric] = scaler.transform(x_v[numeric])
    model = regressor
    model.fit(x_t, y_t)
    predict = model.predict(x_v)
    print('Метод:', info)
    print(f'sMAPE: {sMAPE(predict, y_v):.2f}%')
```

Добавим функцию которая переберет различные методы обработки пропусков, воспользуется предыдущей функцией и вернет результат в читабельном виде

```python
# drop - удалять строки с пропусками в целевом признаке или нет?
# regressor - какой алгоритм использовать для проверки?
# name_model - имя модели для передачи в принт
def check_drop(df, regressor, name_model):
    temp = df.copy()
    temp = temp.iloc[:, 13:36]
    temp.dropna(subset=['rougher.output.recovery'], inplace=True)
    
    print('\033[1m' + '\033[94m' + '='*35 + f'[.....{name_model}.....]' + '='*35 + '\033[0m' + '\n')
    check_on_models(temp.dropna(), 'Полное удаление строк с пропусками', \
                    False, False, False, regressor)
    check_on_models(temp.replace(np.nan, 0), 'Заполнение пропусков нулем', \
                    False, False, False, regressor)
    check_on_models(temp, 'Заполнение пропусков медианой', True, 'median', False, regressor)
    check_on_models(temp, 'Заполнение пропусков средним', True, 'mean', False, regressor)
    check_on_models(temp.ffill(axis=0), 'Заполнение пропусков предыдущим значением по столбцу',\
                False, False, False, regressor)
    check_on_models(temp.dropna(), 'Масштабирование признаков c полным удалением пропусков', \
                    False, False, True, regressor)
```

При помощи функций полностью удалим строки с пропусками у целевого признака и выведем значение sMAPE c 5 различными вариантами заполнения пропусков у остальных признаков, а так же с использованием масштабирования признаков

```python
check_drop(train, LinearRegression(), 'Линейная регрессия')
check_drop(train, DecisionTreeRegressor(random_state=RAST), 'Дерево решений')
check_drop(train, HuberRegressor(), 'Регрессия Хубера')
```

## Модель

Добавим функцию для вычисления итоговой sMAPE. 

```python
def total_sMAPE(best_model_rougher, rougher_name,\
                best_model_final, final_name,\
                X_r_t, y_r_t, X_f_t, y_f_t):
    predict_rougher = best_model_rougher.predict(X_r_t)
    sMAPE_rougher = sMAPE(predict_rougher, y_r_t)
    
    predict_final = best_model_final.predict(X_f_t)
    sMAPE_final = sMAPE(predict_final, y_f_t)
    
    total = 0.25 * sMAPE_rougher + 0.75 * sMAPE_final
    
    result_test = pd.DataFrame(data=[[rougher_name, round(sMAPE_rougher, 3),\
                                      final_name, round(sMAPE_final, 3),\
                                      round(total, 3)]],\
                               columns=['Модель "rougher"','sMAPE, %',\
                                        'Модель "final"','sMAPE, %',\
                                        'Итоговый sMAPE, %'])
    return result_test
```

Для поиска лучшей модели через кросс-валидатор создадим свой scorer при помощи метода make_scorer

```python
smape_score = make_scorer(sMAPE, greater_is_better=False)
```

```python
all_result = pd.DataFrame(columns=['Модель','sMAPE, %', 'Оптимальные параметры',\
                                       'Время выполнения, сек.', 'Эффективность обогащения'])
```

Проведем через pipeline различные модели для предсказания эффективности обогащения чернового концентрата, результаты добавим в таблицу для изучения, все модели с лучшими гиперпараметрами сохраним в отдельные переменные. Для этого напишем необходимую функцию

```python
# Подбор оптимальных параметров модели + вывод результатов + сохранение модели с лучшими параметрами
# regressor - модель
# params - параметры для перебора
# features - признаки
# target - целевой признак
# name_regressor - наименование модели
# pred - наименование эффективности обогащения
def my_pipe(regressor, params, features, target, name_regressor, pred):
    time_start = timeit.default_timer()
    scaler = StandardScaler()
    steps = [('normalizer', scaler), ('classifier', regressor)]
    pipe = Pipeline(steps)
    
    grid_search = GridSearchCV(pipe, params, scoring=smape_score)
    grid_search.fit(features, target)
    
    time = timeit.default_timer() - time_start
    
    result = pd.DataFrame(data=[[name_regressor, abs(round(grid_search.best_score_, 3)),\
                                str(grid_search.best_params_), round(time, 3), pred]],\
                          columns=['Модель','sMAPE, %', 'Оптимальные параметры',\
                                   'Время выполнения, сек.', 'Эффективность обогащения'])
    return result, grid_search.best_estimator_
```

Но все же хочется ради интереса проверить остальные пары моделей на тестовой выборке. Для этого создадим словарь с моделями, прошедшими проверку на адекватность.

```python
rougher_dict = {rougher_LinReg: 'Линейная регрессия',\
                rougher_tree: 'Дерево решений',\
                rougher_OMP: 'Orthagonal Matching Pursuit',\
                rougher_lasso: 'Регрессия Лассо',\
                rougher_NuSVR: 'NuSVR',\
                rougher_random: 'Случайный лес'}

final_dict = {final_tree: 'Дерево решений',\
              final_OMP: 'Orthagonal Matching Pursuit',\
              final_lasso: 'Регрессия Лассо',\
              final_random: 'Случайный лес'}
```

Возможно стоит использовать именно такую пару для предсказаний.
